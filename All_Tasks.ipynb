### PART 1- Randomly Creating Datasets In Broader Ranges - Simpson's Paradox May Or May Not Exist 

The code below generates a random trial in which Simpson's Paradox may or may not exist, and tells you whether Simpson's Paradox exists using if, elif checks.

```python
import numpy as np
from numpy import random

small_stones_a = np.random.randint(10,81)
small_stones_a_group_size = np.random.randint(83,87)
large_stones_a = np.random.randint(10,249)
large_stones_a_group_size = np.random.randint(250,263)

small_stones_b = np.random.randint(10,258)
small_stones_b_group_size = np.random.randint(260,270)
large_stones_b = np.random.randint(10,69)
large_stones_b_group_size = np.random.randint(70,80)


arr = np.array([[small_stones_a, small_stones_a_group_size], #Group 1
                [large_stones_a, large_stones_a_group_size], #Group 3
                [small_stones_b,small_stones_b_group_size], #Group 2
                [large_stones_b,large_stones_b_group_size]]) #Group 4

print("TESTING FOR SIMPSON'S PARADOX")
print("GENERATING RANDOM FOUR DATASETS")
print(arr)


small_stones_a_percentage = small_stones_a / small_stones_a_group_size 
large_stones_a_percentage = large_stones_a / large_stones_a_group_size

small_stones_b_percentage = small_stones_b / small_stones_b_group_size 
large_stones_b_percentage = large_stones_b / large_stones_b_group_size

"""---------------------- Setting up count to analyse whether Simpson's paradox exists -----------------"""

count = 0 

"""---------------------- Calculating The Overall Percent Of Group A ------------------------------------"""
total_a_test_subjects = small_stones_a + large_stones_a
total_a_group_size = small_stones_a_group_size + large_stones_a_group_size
overall_percent_A = total_a_test_subjects / total_a_group_size
overall_A = [total_a_test_subjects,total_a_group_size]
"""---------------------- Calculating The Overall Percent Of Group B ------------------------------------"""


total_b_test_subjects = small_stones_b + large_stones_b
total_b_group_size = small_stones_b_group_size + large_stones_b_group_size
overall_percent_B = total_b_test_subjects / total_b_group_size
overall_B = [total_b_test_subjects,total_b_group_size]

print('\nSMALL STONES TREATMENT')
print('Small stones Treatment A {:.0%} {}'.format(small_stones_a_percentage,arr[0]))
print('Small stones Treatment B {:.0%} {}'.format(small_stones_b_percentage,arr[0]))

if small_stones_a_percentage > small_stones_b_percentage:
    print("------------------------------------")
    print("A is the lead in this clinical trial")
    print("------------------------------------")
    count += 1
else:
    print("------------------------------------")
    print("B is the lead in this clinical trial")
    print("------------------------------------")

    count += 2
    
print('\nLARGE STONES TREATMENT')
print('Large stones Treatment A {:.0%} {}'.format(large_stones_a_percentage,arr[1]))
print('Large stones Treatment B {:.0%} {}'.format(large_stones_b_percentage,arr[1]))

if large_stones_a_percentage > large_stones_b_percentage:
    print("------------------------------------")
    print("A is the lead in this clinical trial")
    print("------------------------------------")
    count += 1
else:
    print("------------------------------------")
    print("B is the lead in this clinical trial")
    print("------------------------------------")

    count += 2


print("\nRESULT WHEN THE GROUPS ARE COMBINED INTO ONE BIG DATA SET")
print('Overall Treatment A {:.0%} {}'.format(overall_percent_A,overall_A))
print('Overall Treatment B {:.0%} {}'.format(overall_percent_B,overall_B))

if overall_percent_A > overall_percent_B:
    print("------------------------------------")
    print("A is the lead in this clinical trial")
    print("------------------------------------")
    count += 1
else:
    print("------------------------------------")
    print("B is the lead overall")
    print("------------------------------------")

    count += 2

if count == 3:
    print("Simpson's Paradox does not exist, A lead in two trials and had overall lead")
elif count == 6:
    print("Simpson's Paradox does not exist, B lead in two trials and had overall lead")
elif count == 4 and small_stones_a_percentage > small_stones_b_percentage and large_stones_a_percentage > large_stones_b_percentage :
    print("Simpson's Paradox exists, A lead in two trials, but B had overall lead")
elif count == 5 and small_stones_b_percentage > small_stones_a_percentage and large_stones_b_percentage > large_stones_a_percentage :
    print("Simpson's Paradox exists, B lead in two trials, but A had overall lead")
```

### PART 2 - Randomly Creating Datasets In Broader Ranges - Simpson's Paradox May Or May Not Exist 
### Letting user set the number of trials and generating a count and bar graph to show number of trials Simpson's Paradox happened

In the code below the user sets the number of random trials to generate in which Simpson's Paradox may or may not exist, and tells you whether Simpson's Paradox exists using if, elif checks, and then displays bar graph of total number of occurrences.

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
Simpson_Paradox_Count = 0

def run_a_trial():
    global number_of_trials
    
    print("TRIAL:" , number_of_trials )
    small_stones_a = np.random.randint(10,81)
    small_stones_a_group_size = np.random.randint(83,87)
    large_stones_a = np.random.randint(10,249)
    large_stones_a_group_size = np.random.randint(250,263)

    small_stones_b = np.random.randint(10,258)
    small_stones_b_group_size = np.random.randint(260,270)
    large_stones_b = np.random.randint(10,69)
    large_stones_b_group_size = np.random.randint(70,80)


    arr = np.array([[small_stones_a, small_stones_a_group_size], #Group 1
                    [large_stones_a, large_stones_a_group_size], #Group 3
                    [small_stones_b,small_stones_b_group_size], #Group 2
                    [large_stones_b,large_stones_b_group_size]]) #Group 4

    print("TESTING FOR SIMPSON'S PARADOX")
    print("GENERATING RANDOM FOUR DATASETS")
    print(arr)


    small_stones_a_percentage = small_stones_a / small_stones_a_group_size 
    large_stones_a_percentage = large_stones_a / large_stones_a_group_size

    small_stones_b_percentage = small_stones_b / small_stones_b_group_size 
    large_stones_b_percentage = large_stones_b / large_stones_b_group_size

    """---------------------- Setting up count to analyse whether Simpson's paradox exists -----------------"""

    count = 0 

    """---------------------- Calculating The Overall Percent Of Group A ------------------------------------"""
    total_a_test_subjects = small_stones_a + large_stones_a
    total_a_group_size = small_stones_a_group_size + large_stones_a_group_size
    overall_percent_A = total_a_test_subjects / total_a_group_size
    overall_A = [total_a_test_subjects,total_a_group_size]
    """---------------------- Calculating The Overall Percent Of Group B ------------------------------------"""


    total_b_test_subjects = small_stones_b + large_stones_b
    total_b_group_size = small_stones_b_group_size + large_stones_b_group_size
    overall_percent_B = total_b_test_subjects / total_b_group_size
    overall_B = [total_b_test_subjects,total_b_group_size]

    print('\nSMALL STONES TREATMENT')
    print('Small stones Treatment A {:.0%} {}'.format(small_stones_a_percentage,arr[0]))
    print('Small stones Treatment B {:.0%} {}'.format(small_stones_b_percentage,arr[0]))

    if small_stones_a_percentage > small_stones_b_percentage:
        print("------------------------------------")
        print("A is the lead in this clinical trial")
        print("------------------------------------")
        count += 1
    else:
        print("------------------------------------")
        print("B is the lead in this clinical trial")
        print("------------------------------------")

        count += 2

    print('\nLARGE STONES TREATMENT')
    print('Large stones Treatment A {:.0%} {}'.format(large_stones_a_percentage,arr[1]))
    print('Large stones Treatment B {:.0%} {}'.format(large_stones_b_percentage,arr[1]))

    if large_stones_a_percentage > large_stones_b_percentage:
        print("------------------------------------")
        print("A is the lead in this clinical trial")
        print("------------------------------------")
        count += 1
    else:
        print("------------------------------------")
        print("B is the lead in this clinical trial")
        print("------------------------------------")

        count += 2


    print("\nRESULT WHEN THE GROUPS ARE COMBINED INTO ONE BIG DATA SET")
    print('Overall Treatment A {:.0%} {}'.format(overall_percent_A,overall_A))
    print('Overall Treatment B {:.0%} {}'.format(overall_percent_B,overall_B))

    if overall_percent_A > overall_percent_B:
        print("------------------------------------")
        print("A is the lead in this clinical trial")
        print("------------------------------------")
        count += 1
    else:
        print("------------------------------------")
        print("B is the lead overall")
        print("------------------------------------")

        count += 2

    global Simpson_Paradox_Count
    if count == 3:
        print("Simpson's Paradox does not exist, A lead in two trials and had overall lead")
    elif count == 6:
        print("Simpson's Paradox does not exist, B lead in two trials and had overall lead")
    elif count == 4 and small_stones_a_percentage > small_stones_b_percentage and large_stones_a_percentage > large_stones_b_percentage :
        print("Simpson's Paradox exists, A lead in two trials, but B had overall lead")
        Simpson_Paradox_Count += 1 
    elif count == 5 and small_stones_b_percentage > small_stones_a_percentage and large_stones_b_percentage > large_stones_a_percentage :
        print("Simpson's Paradox exists, B lead in two trials, but A had overall lead")
        Simpson_Paradox_Count += 1
    else:
        print("Simpson's Paradox does not exist")
    
    return Simpson_Paradox_Count

number_of_trials = 0

choose_number_of_trials = int(input("Write a number for how many trials you want to run:"))

while number_of_trials < choose_number_of_trials:
    number_of_trials += 1
    run_a_trial()

    
print("\n-----------------------------------------------------------------\n")

print("Out of ", choose_number_of_trials,"trials, Simpson's Paradox happened", Simpson_Paradox_Count, "time(s)")

print("\n-----------------------------------------------------------------\n")

Paradox_Did_Not_Happen_Count = choose_number_of_trials - Simpson_Paradox_Count

# https://www.tutorialspoint.com/matplotlib/matplotlib_bar_plot.htm

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
description = ['Paradox Did Not Happen Count', 'Simpson Paradox Count']
data = [Paradox_Did_Not_Happen_Count,Simpson_Paradox_Count]
ax.bar(description,data)
plt.show()
```
